```python
# http://www.nfd.com.tw/lottery/lottyear/year.htm

import pandas as pd

years = range(2004, 2015)

get_hist_table = lambda year: pd.read_html("http://www.nfd.com.tw/lottery/49-year/49-%s.htm" % year)[0][1:]

```

```python
hist_tables = map(get_hist_table,years)
```

```python
df = pd.concat(hist_tables)
```

```python
df
```

```python
df.columns = ["year","monthDay","YN","x1","x2","x3","x4","x5","x6","s","TN"]
```

```python
df = df[["YN","TN","year","monthDay","x1","x2","x3","x4","x5","x6","s"]]
```

```python
df["month"] = df["monthDay"].map(lambda xx:int(xx.split("/")[0]))
```

```python
df["day"] = df["monthDay"].map(lambda xx:int(xx.split("/")[-1]))
```

```python
for col in ["year","x1","x2","x3","x4","x5","x6","s"]:
    df[col] = df[col].map(int)
```

```python
df = df[["TN","YN","year","month","day","x1","x2","x3","x4","x5","x6","s"]]
```
